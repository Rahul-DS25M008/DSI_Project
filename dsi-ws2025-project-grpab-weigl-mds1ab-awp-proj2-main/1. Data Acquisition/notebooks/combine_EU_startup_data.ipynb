# Combine single dataframes

Here we combine the 32 CSVs with all the company data taht was scraped.

```python
import pandas as pd
import glob
import os
```

```python

folder_path = 'EU_startup_data'

all_files = glob.glob(os.path.join(folder_path, "*.csv"))

list_of_dfs = []


for file_path in all_files:
    # Read each CSV 
    df = pd.read_csv(file_path)
    # Append the DataFrame 
    list_of_dfs.append(df)

# Concatenate all DataFrames in the list into one big DataFrame

big_dataframe = pd.concat(list_of_dfs, ignore_index=True)

print(f"Total number of files found: {big_dataframe.shape}")
```

```python
big_dataframe.duplicated().sum()
```

As expected there are no duplicates in the datframe.

```python
# savee the combined dataframe

big_dataframe.to_csv("EU_startups_full.csv", index=False)
```
