```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../1. Data Acquisition/data/EU_startups_full.csv")
```

```python
df.head(10)
```

```python
df.rename(columns={
    'company name': 'company_name',
    'category': 'country',
    'based in': 'city' ,
    'total funding' : 'funding_amount',
    'tags' : 'industry_tags',
    'founded':'start_year',
    'website':'website',
    'company status':'status'
}, inplace=True)
```

```python
df.head(5)
```

```python
df['funding_amount'].unique()
```

```python
replacement_map = {
    'No funding announced yet': np.nan,
    'Between €1-€100K': 50000,
    'Between €100K-€500K': 300000,
    'Between €500K-€ 1 million': 750000,
    'Between €1 million-€ 2.5 million' : 1750000,
    'Between €2.5 million-5 million' : 3750000,
    'Between €5 million-€10 million' : 7500000,
    'Between €10 million-€25 million' : 17500000,
    'Above €25 million' : 25000000
}
```

```python
df['funding_amount'] = df['funding_amount'].replace(replacement_map)
```

```python
df.head(10)
```

```python
df['status'].unique()
```

```python
replacement_map_status = {
    'Active': True,
    'Idea Stage': True,
    'Closed': False
}
```

```python
df['status'] = df['status'].replace(replacement_map_status)
```

```python
df.head(10)
```

```python
df['status'] = df['status'].astype('bool')
```

```python
print(df.dtypes)
```

```python
df = df[["country", "company_name", "city", "funding_amount", "industry_tags", "start_year", "status", "website"]]
```

```python
df.head(5)
```

```python
df.to_csv("../data/eu_startups.csv", index=False)
```
