### Filter for Europe in OECD

```python
import pandas as pd
```

```python
df_oecd =  pd.read_csv("../../1. Data Acquisition/data/oecd_data.csv")
```

```python
df_oecd["Reference area"].unique()
```

```python
countries_to_drop = ['Australia', 'Canada', 'Israel','Japan', 'Korea',  'New Zealand', 'Russia' , 'United States',
       'South Africa']
```

```python
rows_to_keep = ~df_oecd["Reference area"].isin(countries_to_drop)

df_filtered = df_oecd[rows_to_keep]
```

```python
df_filtered["Reference area"].unique()
```

```python
df_filtered["REF_AREA"].unique()
```

```python
df_filtered.head()
```

```python
new_cols = ["country_code", "country", "business_dev_satge", "year", "amount"]
```

```python
df_filtered.columns = new_cols
```

```python
df_filtered.head()
```

```python
df_filtered.to_csv('../data/OECD_eu_investment.csv', index=False)
```
